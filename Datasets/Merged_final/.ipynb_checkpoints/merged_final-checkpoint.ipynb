```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
snap = pd.read_csv('../SNAP/ACS_09-15.csv')
```

```python
snap.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
snap_0914 = snap[snap['Year'] != 2015]
```

```python
snap_15 = snap[snap['Year'] == 2015]
```

```python
snap_0914 = snap_0914[['id', 'id2', 'HC01_EST_VC01', 'HC02_EST_VC01', 'Year']]
```

```python
snap_0914.rename(columns={'id': 'GEO.id', 'id2': 'GEO.id2', 'Year': 'year', 
                          'HC01_EST_VC01': 'Household_total', 'HC02_EST_VC01': 'Household_Food_Stamp_Received'}, inplace=True)
```

```python
snap_15 = snap_15[['id', 'id2', 'HC01_EST_VC01', 'HC03_EST_VC01', 'Year']]
```

```python
snap_15.rename(columns={'id': 'GEO.id', 'id2': 'GEO.id2', 'Year': 'year', 
                        'HC01_EST_VC01': 'Household_total', 'HC03_EST_VC01': 'Household_Food_Stamp_Received'}, inplace=True)
```

```python
snap_merged = snap_0914.append(snap_15)
```

```python
income = pd.read_csv('../Income_Language_Race/income.csv')
income.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
race = pd.read_csv('../Income_Language_Race/race.csv')
race.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
language_no09 = pd.read_csv('../Income_Language_Race/language_no09data.csv')
language_no09.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
income.columns
```

```python
income.rename(columns={'HC01_EST_VC02': 'Income_Less_than_$10,000', 
                       'HC01_EST_VC03': 'Income_$10,000_14,999', 
                       'HC01_EST_VC04': 'Income_$15,000_24,999', 
                       'HC01_EST_VC05': 'Income_$25,000_34,999', 
                       'HC01_EST_VC06': 'Income_$35,000_49,999', 
                       'HC01_EST_VC07': 'Income_$50,000_74,999', 
                       'HC01_EST_VC08': 'Income_$75,000_99,999', 
                       'HC01_EST_VC09': 'Income_$100,000_149,999', 
                       'HC01_EST_VC10': 'Income_$150,000_199,999', 
                       'HC01_EST_VC11': 'Income_$200,000_or_more', 
                       'HC01_EST_VC13': 'Median_income'}, inplace=True)
```

```python
for column in race.columns[3:-1]:
    race[column] = race[column]/(race['HD01_VD01'].astype(float))
```

```python
race.drop('HD01_VD01', axis=1, inplace=True)
```

```python
race.rename(columns={'HD01_VD02': 'White_alone_percent',
                     'HD01_VD03': 'Black_or_African_American_alone_percent',
                     'HD01_VD04': 'American_Indian_and_Alaska_Native_alone_percent',
                     'HD01_VD05': 'Asian_alone_percent',
                     'HD01_VD06': 'Native_Hawaiian_and_Other_Pacific_Islander_alone_percent',
                     'HD01_VD07': 'Some_other_race_alone_percent',
                     'HD01_VD08': 'Two_or_more_races_percent',
                     'HD01_VD09': 'Two_races_including_Some_other_race_percent',
                     'HD01_VD10': 'Two_races_excluding_Some_other_race_and_three_or_more_races_percent'}, inplace=True)
```

```python
snap_merged.rename(columns={'id': 'GEO.id', 'id2': 'GEO.id2', 'Year': 'year', 
                            'HC01_EST_VC01': 'Household_total', 'HC02_EST_VC01': 'Household_Food_Stamp_Received'}, inplace=True)
```

```python
pd.merge(snap_merged, income, on=['GEO.id', 'GEO.id2', 'year'])
```

```python
# HC01_EST_VC02: Speak only English 
# HC01_EST_VC03: Speak a language other than English 
# HC01_EST_VC04: SPEAK A LANGUAGE OTHER THAN ENGLISH - Spanish 
# HC01_EST_VC05: SPEAK A LANGUAGE OTHER THAN ENGLISH - Other Indo-European languages
# HC01_EST_VC06: SPEAK A LANGUAGE OTHER THAN ENGLISH - Asian and Pacific Island languages 
# HC01_EST_VC07: SPEAK A LANGUAGE OTHER THAN ENGLISH - Other languages
```

```python
language_no09.rename(columns={'HC01_EST_VC02': 'Speak only English', 
                              'HC01_EST_VC03': 'Speak a language other than English', 
                              'HC01_EST_VC04': 'SPEAK A LANGUAGE OTHER THAN ENGLISH - Spanish', 
                              'HC01_EST_VC05': 'SPEAK A LANGUAGE OTHER THAN ENGLISH - Other Indo-European languages',
                              'HC01_EST_VC06': 'SPEAK A LANGUAGE OTHER THAN ENGLISH - Asian and Pacific Island languages',
                              'HC01_EST_VC07': 'SPEAK A LANGUAGE OTHER THAN ENGLISH - Other languages'}, inplace=True)
```

```python
snap_merged_income = pd.merge(snap_merged, income, on=['GEO.id', 'GEO.id2', 'year'])
```

```python
snap_merged_race_income = pd.merge(snap_merged_income, race, on=['GEO.id', 'GEO.id2', 'year'])
```

```python
snap_merged_race_income.to_csv('snap_merged_race_income_09_14.csv')
```
