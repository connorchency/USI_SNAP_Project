```python
def closest_retailer(ct_centroid_df, retailerNYC):
    dist = {}
    for i, row_ct in ct_centroid_df.iterrows():
        dist[row_ct['Id']] = 9999
        for j, row_re in retailerNYC.iterrows():
            distance = vincenty((row_ct['geometry'].y, row_ct['geometry'].x), 
                                (row_re['geometry'].y, row_re['geometry'].x)).miles
            if distance < dist[row_ct['Id']]:
                dist[row_ct['Id']] = distance
    return dist
```
